```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
file_path = 'https://raw.githubusercontent.com/Heytec/Africa-Data-School-Curriculum/main/notebooks/data/jumta.csv'
jumta = pd.read_csv(file_path)
jumta
```

```python
plt.figure(figsize=(15,5))
plt.plot(jumta['Month'],jumta['Revenue'], color = 'green', linestyle ='-', marker='o',linewidth= 2, label= 'Revenue')
plt.title('Monthly Revenue 2020-2021', color='red')
plt.xlabel('Months')
plt.ylabel('Revenue')
plt.xticks(rotation=45)
plt.grid()
plt.legend()

plt.show()
```

```python
plt.figure(figsize=(15,5))
plt.bar(jumta['Month'],jumta['Orders'], color='blue', label='Orders')
plt.title('Total Monthly Orders 2020-2021', color='Red')
plt.xlabel('Months')
plt.ylabel('Monthly Orders')
plt.xticks(rotation=45)
plt.legend()
plt.grid(axis='y')
plt.show
```

```python
plt.figure(figsize=(10,5))
plt.scatter(jumta['Orders'],jumta['Average_Order_Value'], color = 'red', label='Avg order value')
plt.title('Number of Orders vs Average Order Value 2020-2021')
plt.xlabel('Orders')
plt.ylabel('Average Order Value')
plt.legend()
plt.grid()
```

```python
fig, (ax1, ax2, ax3) = plt.subplots(nrows=3, ncols=1, figsize=(10,15))

ax1.plot(jumta['Month'],jumta['Revenue'], color = 'green', linestyle ='-', marker='o',linewidth= 2, label= 'Revenue')
ax1.set_title('Monthly Revenue 2020-2021', color='red')
ax1.set_xlabel('Months')
ax1.set_ylabel('Revenue')
ax1.tick_params(axis='x',rotation=45)
ax1.grid()
ax1.legend()

ax2.bar(jumta['Month'],jumta['Orders'], color='blue', label='Orders')
ax2.set_title('Total Monthly Orders 2020-2021', color='Red')
ax2.set_xlabel('Months')
ax2.set_ylabel('Monthly Orders')
ax2.tick_params(axis= 'x', rotation=45)
ax2.legend()
ax2.grid(axis='y')

ax3.scatter(jumta['Orders'],jumta['Average_Order_Value'], color = 'red', label='Avg order value')
ax3.set_title('Number of Orders vs Average Order Value 2020-2021')
ax3.set_xlabel('Orders')
ax3.set_ylabel('Average Order Value')
ax3.legend()
ax3.grid()

plt.tight_layout()
plt.savefig('Jumta_Sales_data_analysis.png', dpi=300)
plt.show()
```
