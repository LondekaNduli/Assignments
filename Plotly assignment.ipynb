```python
import plotly.express as px
import pandas as pd
```

```python
titanic=pd.read_csv('https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv')
titanic
```

```python
fig=px.histogram(titanic,x='Age', y='Survived',color='Sex', title='Passengers who survived Categorised by Age',color_discrete_sequence=['purple', 'hotpink'])
fig.show()
```

```python
fig=px.violin(titanic,x='Pclass', y='Fare',color='Pclass', title=' Distribution of Fare prices by Passenger Class',
              color_discrete_sequence=['purple', 'hotpink', 'orange'],
             labels={'Fare':'Fare Price', 'Pclass':'Passenger Class'}, box=True)
fig.show()
```

```python
fig=px.scatter(titanic,x='Age', y='Fare',color='Pclass', title=' Distribution of Fare prices by Passengers` Ages',
              color_discrete_sequence=['purple', 'hotpink', 'orange'],
             labels={'Fare':'Fare Price', 'Age':'Passenger Age', 'Pclass':'Passenger Class'})
fig.show()
```

```python
fig=px.scatter_matrix(titanic,dimensions=['Age','Fare','Pclass'],color='Survived', 
                      title='Scatter Plot Matrix of Age, Fare, and Class by Survival of Passengers',
              color_discrete_sequence=['purple', 'hotpink',],
             labels={'Fare':'Fare Price', 'Age':'Passenger Age', 'Pclass':'Passenger Class'})
fig.show()
```

```python
import plotly.io as pio

pio.write_html(fig, 'bar_plot.html')
pio.write_html(fig, 'histogram_plot.html')
pio.write_html(fig, 'violin_plot.html')
pio.write_html(fig, 'scatter_plot.html')
pio.write_html(fig, 'scatter_matrix.html')
```
